```python
df_new = df[['VendorID','passenger_count','trip_distance','total_amount','tip_amount','payment_type']]
df_new
```

```python
## Sort the data by trip distance from maximum to minimum value
df_sort= df_new.sort_values(by=['trip_distance'],ascending=False)
df_sort.head(10)
```

```python
# Sort the data by total amount and print the top 20 values
total_amount_desc = df_new.sort_values(by=['total_amount'],ascending=False)['total_amount']
total_amount_desc.head(20)
```

```python
# Sort the data by total amount and print the bottom 20 values
total_amount_desc.tail(20)
```

```python
# How many of each payment type are represented in the data?
payment_methods = df_new['payment_type'].value_counts()
payment_methods
```

```python
# What is the average tip for trips paid for with credit card?
avg_tips = df_new[df['payment_type']==1] ['tip_amount'].mean()
avg_tips
```

```python
# What is the average tip for trips paid for with cash?
avg_tips = df_new[df['payment_type']==2] ['tip_amount'].mean()
avg_tips
```

```python
# What is the mean total amount for each vendor?
df_new.groupby(['VendorID']).mean(numeric_only=True) [['total_amount']]
```

```python
# Filter the data for credit card payments only
mask = df_new['payment_type'] == 1
mask

credit_card = df_new[mask]
credit_card['passenger_count'].value_counts()
```

```python
average_tip_passenger_count= credit_card.groupby(['passenger_count']).agg(['mean'])[['tip_amount']]

# Calculate the average tip amount for each passenger count (credit card payments only)

average_tip_passenger_count
```
